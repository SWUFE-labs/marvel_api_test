```python
with open('hero_dict_n.txt','r') as hero_dict:
    
    hero_dict_j = json.loads(hero_dict.read())
    name_l=[]
    event_l=[]
    comic_l=[]
    story_l=[]
    for ids in hero_dict_j['heros_list']:
        id_ = ids['id']
        name = ids['name']    
        comics = int(ids['comics'])
        events = int(ids['events'])
        stories = int(ids['stories'])
        if events >= 0:
            #print(id_,name,events,comics,stories)
            event_l.append(events)
            name_l.append(name)
            comic_l.append(comics)
            story_l.append(stories)
```

```python
#pyy.init_notebook_mode(connected=True)
events = go.Bar(
    x=name_l,
    y=event_l,
    name='Events'
)
comics = go.Bar(
    x=name_l,
    y=comic_l,
    name='Comics'
)
stories = go.Bar(
    x=name_l,y=story_l,name='Stories')
data = [events, comics,stories]

pyy.iplot(data)
```
