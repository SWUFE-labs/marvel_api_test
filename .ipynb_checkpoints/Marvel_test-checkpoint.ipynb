```python
#首先把需要用的库引入进来
from urllib import request  #用于请求网页内容
from IPython.display import Image  #用于显示图片
import time  #用于生成时间戳
from hashlib import md5 #用于生成MD5值
import json #用于处理JSON格式
import pprint #用于输出更容易查看的字典数据
```

```python
#获取Marvel所有英雄的基本数据（API）
#Marvel API的基本调用格式要求：
#http://gateway.marvel.com/v1/public/characters?ts=1&apikey=1234&hash=ffd275c5130566a2916217b101f26150
#其中ts是需要在每次访问的时候提供的时间戳，每次不一样。apikey是注册账号后获得的public key。
#hash是对由ts+private key+public key构成的字符串进行MD5运算后得到的结果，是一个16进制的数字，需转换为字符串使用。
#以下为我的账号相对应的public key 和 private key，需要替换为你们自己的。
#public key: xxxxx 
#private key: xxxxxx
```

# <font color=red>运行以前将pukey和pvkey改成你们自己的！！!</font>

```python
pbkey = '这里替换为你们自己的public key'
pvkey = '这里替换为你们自己的private key'
ts = str(time.time())[-6:]   #生成每次访问的时间戳，只取后六位。可以试下全部都取会有什么问题。
```

```python
def hashkey(ts,pvkey,pbkey):    #生成hash值的函数
    pvk = pvkey
    pbk = pbkey
    key = ts + pvk + pbk
    _md5key = str(md5(key.encode('utf-8')).hexdigest())
    return _md5key

md5key = hashkey(ts,pvkey,pbkey)

name = 'spider-man'  #设定要查找的英雄的名字的开头部分
limit = '5' #限制每次返回结果的数量，有效值为1-100之间。建议调试程序的时候这个值小一点。

#生成api的调用链接
url_allheros = "https://gateway.marvel.com:443/v1/public/characters?nameStartsWith="+name+"&limit="+limit+"&ts="+ts+"&apikey="+pbkey+"&hash="+md5key
#打印出来检验一下
print(url_allheros)

#获取数据
all_heros = request.urlopen(url_allheros)
```

```python
#将取得的原始数据分离出来
ah = all_heros.read()
```

```python
#原始数据是JSON格式的，在python中需要转换为字典数据类型才能正常使用
all_heros_json = json.loads(ah)
```

```python
#pprint.pprint(all_heros_json)  #用pprint生成更容易读懂的字典格式，方便寻找需要的关键字和值

#通过all_heros_json里的count关键字获取本次取得了几条有效数据
print(all_heros_json['data']['count'],'items succeed!')
```

```python
#将英雄的相关信息从逐条数据中取出并输出
for hero in all_heros_json['data']['results']:
    print('[Hero Name]: ',hero['name'],'\n[MARVEL ID]:',hero['id'])
    des = hero['description']
    if des == '':    #description有些是空值，为了方便阅读，把空值替换为None
        des = 'None'
    print('[Description]:\n',des)
    img = hero['thumbnail']['path']+'.'+hero['thumbnail']['extension']
    display(Image(img,width=400))
    print('Photo: ',img,'\n','---------------')
   
```
