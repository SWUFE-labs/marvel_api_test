```python
# while True:   
#     name = input('Search for : ')
#     if name == '':
#         continue
#     break
    
# limit = input('Results limited to (1-100, 1 for defualt) :')
# try:
#     int(limit)
# except:
#     limit = "1"

limit = '100'
offset = 0
count = 1
id_list =[]

while count > 0 :
    md5key,ts = marvel.hashkey()
    url =  "https://gateway.marvel.com:443/v1/public/characters?"\
                      "offset="+str(offset)+"&limit="+limit+"&ts="+ts+"&apikey="+marvel.pbkey+"&hash="+md5key
    print('['+time.ctime()+']','fetching',offset,'-',offset+100,'......',end='')
    heros = requests.get(url)
    if heros.status_code == 200:
        print('[Done]')
    print('['+time.ctime()+']','Decoding.....',end='')
    heros_r = json.loads(heros.text)
    print('[Done]')
    count = heros_r['data']['count']
    #print(count)
    offset += 100
    print('['+time.ctime()+']','Appending to Dict......',end='')
    for hero in heros_r['data']['results']:
        id_list.append(dict({'id':str(hero['id']),'name':hero['name'],'events':hero['events']['available'],
                            'comics':hero['comics']['available'],'series':hero['series']['available'],
                            'stories':hero['stories']['available'],'description':hero['description'],
                            'modified':hero['modified'],'thumbnail':hero['thumbnail']}))     
    print('[Done]')

hero_dict={'heros_list':id_list}
print('['+time.ctime()+']','Converting to JSON ......',end='')
hero_j = json.dumps(hero_dict)
print('[Done]')

print('['+time.ctime()+']','Writting to',os.getcwd()+'/hero_dict_n.txt ......',end='')
with open('hero_dict_n.txt','w') as hero_f:
    hero_f.write(str(hero_j))
print('[Done]')
```
