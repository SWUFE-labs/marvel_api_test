```python
#首先把需要用的库引入进来
import requests
from IPython.display import Image
import time
from hashlib import md5
import json
import os
```

```python
#获取Marvel所有英雄的基本数据（API）
#http://gateway.marvel.com/v1/public/comics?ts=1&apikey=1234&hash=ffd275c5130566a2916217b101f26150
#public key: b429b36ad9c859671f28487a8acdca22 
#private key: 10486b67a96262c7c4859f963ae11c36d94352e6

pbkey = 'b429b36ad9c859671f28487a8acdca22'
pvkey = '10486b67a96262c7c4859f963ae11c36d94352e6'

#定义md5key生成函数
def hashkey(pvkey,pbkey):
    ts = str(time.time())[-6:]   #timestamp ts
    key = ts + pvkey + pbkey
    _md5key = md5(key.encode('utf-8')).hexdigest()
    return _md5key,ts
```

```python
# while True:   
#     name = input('Search for : ')
#     if name == '':
#         continue
#     break
    
# limit = input('Results limited to (1-100, 1 for defualt) :')
# try:
#     int(limit)
# except:
#     limit = "1"

limit = '100'
offset = 0
count = 1
hero_dict={}
while count > 0 :
    md5key,ts = hashkey(pvkey,pbkey)
    url =  "https://gateway.marvel.com:443/v1/public/characters?"\
                      "offset="+str(offset)+"&limit="+limit+"&ts="+ts+"&apikey="+pbkey+"&hash="+md5key
    print('['+time.ctime()+']','fetching',offset,'-',offset+100,'......',end='')
    heros = requests.get(url)
    if heros.status_code == 200:
        print('[Done]')
    print('['+time.ctime()+']','Decoding.....',end='')
    heros_r = json.loads(heros.text)
    print('[Done]')
    count = heros_r['data']['count']
    #print(count)
    offset += 100
    print('['+time.ctime()+']','Appending to Dict......',end='')
    for hero in heros_r['data']['results']:
        hero_dict[str(hero['id'])] = hero['name']
    print('[Done]')

print('['+time.ctime()+']','Converting to JSON ......',end='')
hero_j = json.dumps(hero_dict)
print('[Done]')

print('['+time.ctime()+']','Writting to',os.getcwd()+'/hero_dict.txt ......',end='')
with open('hero_dict.txt','w') as hero_f:
    hero_f.write(str(hero_j))
print('[Done]')
```
